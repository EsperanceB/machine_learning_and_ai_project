```python
import pandas as pd
import json

# Load JSON data
with open('HistoricalData.json') as file:
    data = json.load(file)

# Convert to DataFrame
df = pd.DataFrame(data)

# Export to CSV
df.to_csv('ClimateModelingDataset.csv', index=False)
```
